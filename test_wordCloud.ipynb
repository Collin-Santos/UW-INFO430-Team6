```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_csv("tweets.csv")
df.head()
```

```python
text = df['content']
text.head()
```

```python
comment_word = ''
stopwords = set(STOPWORDS)
```

```python
for val in df.content: 
    val = str(val)
    tokens = val.split()

    #tokens = re.sub(r'\W+', '', val).lower()
    
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        tokens[i] = tokens[i].replace("https","")
        tokens[i] = tokens[i].replace("t","")
        tokens[i] = tokens[i].replace("co","")
        #tokens[i] = tokens[i].replace("s","")
        #tokens[i] = tokens[i].replace("et","")
    #print(tokens)
    comment_word += " ".join(tokens)+" "
    print(tokens)
```

```python
wordCloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_word)
```

```python
plt.figure(figsize = (5, 5), facecolor = None)
plt.imshow(wordCloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
#plt.show()
plt.savefig('wordCloud.png')
```
